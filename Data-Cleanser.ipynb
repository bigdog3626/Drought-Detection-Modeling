```python
import pandas as pd
import os
import numpy as np
```

```python
data_path = 'Data/Little-River-SCAN-Data'
frames = []


for filename in os.listdir(data_path):
    file = os.path.join(data_path, filename)
    df = pd.read_csv(file, skiprows=4)
    df.name = filename[-8:-4]
    max_number_of_nas = 360
    df.replace(float(-99.9), np.NaN, inplace=True)
    df = df.loc[:, (df.isnull().sum(axis=0) < max_number_of_nas)]
    frames.append(df)
```

```python

def check_errors(original_df, df1):
    original_columns = original_df.columns
    columns1 = df1.columns

    common_cols = original_columns.intersection(columns1)
    first_not_next = original_columns.difference(columns1)

    print(f'Columns in Both: {common_cols}')
    print(f'Columns in {original_df.name} but not in {df1.name}: {first_not_next}')
```
